```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import math
csv = "/content/drive/MyDrive/amazon_dataset.csv"
df = pd.read_csv(csv)
df.head(5)   #show only top 5
```

```python
data = df.copy()  # copy all things from df to data variable
data.describe()   #The describe() method is used for calculating some statistical data like percentile, mean and std of the numerical values of the Series or DataFrame. It analyzes both numeric and object series and also the DataFrame column sets of mixed data types.
```

```python
print(len(data))  # normally dekhale amn dilei hoy
print("Before {}".format(len(data))) # .format ar vitor na rakle error dekhabe bcz amra akta string hisebe before ar sathe dekhaci
dataAfter = data.dropna(subset=["reviews.rating"]) # removes all NAN in reviews.rating
print("After {}".format(len(dataAfter)))
dataAfter["reviews.rating"] = dataAfter["reviews.rating"].astype(int)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
s = StratifiedShuffleSplit(test_size=0.3)  #akhan e test_size=0.3 mane total data r 30% se test korar jonno rakbe ar baki 70% data diye se machine ke bujhabe je ki holo kmn hoa dorker output 
for train_index, test_index in s.split(dataAfter, dataAfter["reviews.rating"]): #split diye alada kora hoyeche (baki sob info , only rating)
    strat_train = dataAfter.reindex(train_index)
    strat_test = dataAfter.reindex(test_index)

#strat_test

strat_train
```

```python
def sentiments(rating):
    
    if (rating == 5) or (rating == 4):
        return "Positive"
        
    elif rating == 3:
        return "Neutral"

    elif (rating == 2) or (rating == 1):
        return "Negative"
# Add sentiments to the data
strat_train["Sentiment"] = strat_train["reviews.rating"].apply(sentiments)
strat_test["Sentiment"] = strat_test["reviews.rating"].apply(sentiments)
strat_train["Sentiment"][:50]
#strat_test["Sentiment"][:50]
```
